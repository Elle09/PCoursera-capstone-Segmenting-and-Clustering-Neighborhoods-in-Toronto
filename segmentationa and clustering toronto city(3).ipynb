```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
import requests 
import lxml
from bs4 import BeautifulSoup 
import wikipedia as wp

print('Libraries imported.')
```

```python
#Scraping the Wikipedia page for the table of postal codes of Canada
```

```python
my_url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
r=requests.get(my_url)
#After retreiving the URL and creating a Beautiful soup object** 

soup=BeautifulSoup(r.text,"html.parser")
table = soup.table
table = soup.find("table")
table_rows = table.tbody.find_all("tr")
#finding the table and table data 
print('Page Scrapped.')
```

```python
#The html table is converted to Pandas DataFrame for cleaning and preprocessing.
df = pd.read_csv('Neighborhoods in Toronto.csv',
                    usecols = ['PostalCode', 'Borough', 'Neighborhood'])
df
```

```python
#Data preprocessing and cleaning
```

```python
df1 = df[df.Borough != 'Not assigned']

# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

```python
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values

# libraries for displaying images
from IPython.display import Image 
from IPython.core.display import HTML 


from IPython.display import display_html
```

```python
#Importing the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada
```

```python

lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')

lat_lon.columns = ['PostalCode', 'Latitude', 'Longitude']
lat_lon.head()
```

```python
#Merging the two tables for getting the Latitudes and Longitudes for various neighbourhoods in Canada
df3 = pd.merge(df2,lat_lon,on='PostalCode')
df3.head()
```

```python
#Clustering and the plotting of the neighbourhoods of Canada which contain Toronto in their Borough
```

```python

df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans


import folium # map rendering library

print('Libraries imported.')
```

```python
#Visualizing all the Neighbourhoods of the above data frame using Folium
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
##Using KMeans clustering for the clsutering of the neighbourhoods
```

```python


k=5
toronto_clustering = df4.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4

```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
